```python
import requests
import json
import prettytable
from Series_ID import*
import pandas as pd


```

```python
headers = {'Content-type': 'application/json'}
s_id = [total_nonfarm, construction,edu_health,mining_logging,fin_act,gov,info,leisure_hospi,manu,other_ser,pro_bus,loc_gov,tran_util]
data = json.dumps({"seriesid":s_id,"startyear":"2008", "endyear":"2017"})

p = requests.post('https://api.bls.gov/publicAPI/v2/timeseries/data/', data=data, headers=headers)
json_data = json.loads(p.text)
for series in json_data['Results']['series']:
    x=prettytable.PrettyTable(["series id","year","period","value"])
    seriesId = series['seriesID']
    for item in series['data']:
        year = item['year']
        period = item['period']
        value = item['value']
        if 'M01' <= period <= 'M12':
            x.add_row([seriesId,year,period,value])
    output = open(seriesId + '.txt','w')
    output.write (x.get_string())
    output.close()
```
