```python
import pandas as pd
import matplotlib as plt
import numpy as py
import os
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
```

```python
#read excel file, create df
data_sectorsdf= pd.read_excel("data_for_barchart.xlsx")
data_sectorsdf

```

```python

data_sectorsdf.dropna()                              
               
```

```python
# further re-arrange the column, only keeping(Industry Name, 2006, 2016, total employment)
industry_06=data_sectorsdf.drop(["Year_2016","change_06_16","Percent Distribution","Compound Annual Rate of Change","Unnamed: 6"], axis = 1)
industry_06.dropna()


```

```python
# further plotting the data, in pie chart to see total % of people employed in all three sectors.

labels = ["Goods-producing industry","Services industries","Agriculture Industry"]
sizes = [22466.7,114724.2,2111.2]
colors = ["gold", "blue", "coral"]

explode = (0,0,0.1)

plt.pie(sizes, labels=labels, explode = explode, autopct="%1.1f%%",colors=colors, counterclock=True, startangle=140)
plt.title("% people employed in 2006")
plt.axis("equal")

plt.savefig("2006pie.png")
plt.show()
```
